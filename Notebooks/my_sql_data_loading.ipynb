```python
import mysql.connector
```

```python
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="password",  # put your MySQL password
    database="sales_analysis"
)

cursor = conn.cursor()
```

```python
import pandas as pd

df = pd.read_csv("sales_analysis.csv")
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df.columns
```

```python
df = df[[
    'order_id',
    'order_date',
    'product_id',
    'product_category',
    'customer_region',
    'payment_method',
    'quantity_sold',
    'price',
    'discount_percent',
    'total_revenue'
]]
```

```python
data = [tuple(x) for x in df.to_numpy()]

cursor.executemany("""
    INSERT INTO sales (
        order_id, order_date, product_name, category, region,
        payment_method, quantity_sold, price, discount, total_revenue
    )
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
""", data)

conn.commit()
```
